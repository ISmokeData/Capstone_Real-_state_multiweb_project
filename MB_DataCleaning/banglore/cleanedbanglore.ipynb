# Extract Cr and Lac 

```python
def extract_values(value):
    matches = re.findall(r'(\d+(\.\d+)?\s*Cr|\d+(\.\d+)?\s*Lac)', value)
    return ', '.join([match[0] for match in matches])
```

```python
price = df['price_sqft_price'].apply(extract_values)
```
