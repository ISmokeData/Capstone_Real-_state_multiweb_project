```python
with open('hydrabad.html','r',encoding = 'utf-8') as f:
  html = f.read()
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'lxml')
```

```python
soup.find_all('h2',{'class':'mb-srp__card--title'})
```

```python
container1=soup.find_all('h2',{'class':'mb-srp__card--title'})
```

```python
bhk = []
```

```python
for tag in container1:
  try:
    bhk.append(tag.get_text(strip=True))
  except:
    bhk.append('NAN')
```

```python
bhk
```

```python
len(soup.find_all('div',{'class':'mb-srp__card__summary--value'}))
```

```python
summary_lists = soup.find_all('div', class_='mb-srp__card__summary__list')

properties = []
for summary_list in summary_lists:
    property_details = {}
    items = summary_list.find_all('div', class_='mb-srp__card__summary__list--item')
    for item in items:
        label = item.find('div', class_='mb-srp__card__summary--label').get_text(strip=True)
        value = item.find('div', class_='mb-srp__card__summary--value').get_text(strip=True)
        property_details[label] = value
    properties.append(property_details)

# Print the grouped property details
for i, property_details in enumerate(properties, start=1):
    print(f'Property {i}:')
    for label, value in property_details.items():
        print(f'  {label}: {value}')
```

```python
price_divs = soup.find_all('div',{'class':'mb-srp__card__price'})
```

```python
price_sqft_price = []
prices = []
for price_div in price_divs:
    price_text = price_div.get_text(strip=True).replace('', '').strip()
    prices.append(price_text)

# Print the extracted prices
for i, price in enumerate(prices, start=1):
    price_sqft_price.append(price)
    print(price)
```

```python
def dict_to_string(d):
    if isinstance(d, dict):
        return ", ".join([f"{k}: {v}" for k, v in d.items()])
    else:
        return str(d)
```
